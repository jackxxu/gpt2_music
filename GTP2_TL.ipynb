
## download the model


```
model_name = "124M"
if not os.path.isdir(os.path.join("models", model_name)):
	print(f"Downloading {model_name} model...")
	gpt2.download_gpt2(model_name=model_name)   # model is saved into current directory under /models/124M/
sess = gpt2.start_tf_sess()
```

## train the model

```
!wget --no-check-certificate 'https://docs.google.com/uc?export=download&id=1qyunayxO_xZdPOjZfK_OciWiPUu4Qurp' -O songs.csv
file_name = 'songs.csv'
gpt2.finetune(sess, file_name, model_name=model_name, steps=200)   # steps is max number of training steps
```
